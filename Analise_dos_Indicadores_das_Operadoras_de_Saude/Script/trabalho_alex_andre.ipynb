###Funções

```
#definição de algumas funções
#função para calcular a média
def media(var):
  soma=0
  #estrutura de repetição para somar os valores da variável que vai calcular a média e porteriormente dividir essa soma e achar a média
  for i in range(len(dados1)):
    soma = soma + dados1[var][i]
  media = soma/len(dados1)
  return(media)
#função para calcular a mediana
def mediana(var):
  #criação de uma lista com os valores da variável selecionada em ordem crescente 
  ordem=sorted(dados1[var])
  #estrutura condicional para verificar se o número de dados da variável é par ou ímpar e à partir disso encontrar a mediana
  if len(ordem) % 2 == 0:
    mediana=(ordem[int(len(ordem)/2)]+ordem[int((len(ordem)+2)/2)])/2
  else:
    mediana=ordem[int((len(ordem)+1)/2)]
  return(mediana)
#função para calcular a variância
def variancia(var):
  soma = 0
  #estrutura de repetição para calcular a soma dos dados da variável selecionada subtraindo a média da variável e elevando o valor ao quadrado para posteriormente calcular a variância
  for i in range(len(dados1)):
    soma += (dados1[var][i] - media(var))**2
  variancia=soma/(len(dados1)-1)
  return(variancia)
#função para calcular a covariância
def cov(var1,var2):
  soma=0
  #estrutura de repetição para calcular a soma dos dados da variável 1 subtraindo sua média e multiplicando pelos dados da variável 2 subtraindo sua média para posteriormente calcular o coeficiente
  for i in range(len(dados1)):
    soma += ((dados1[var1][i]-media(var1))*(dados1[var2][i]-media(var2)))
  cov=soma/(len(dados1)-1)
  return(cov)
#função para calcular o coeficiente de correlação de pearson
def ccp(var1,var2):
  ccp=cov(var1,var2)/((variancia(var1)*variancia(var2))**(1/2))
  return(ccp)
#função para ver qual variável o usuário escolheu e transformar para o jeito que ela está escrita no conjunto de dados
def var(x):
  #estrutura condicional para verificar o valor inserido pelo usuário
  if x==1:
    return ('ativo_total')
  elif x==2:
    return ('ativo_circ')
  elif x==3:
    return ('ativo_nao')
  elif x==4:
    return ('passivo_total')
  elif x==5:
    return ('passivo_circ')
  elif x==6:
    return ('passivo_nao')
  elif x==7:
    return ('receitas_total')
  elif x==8:
    return ('despesas_total')
  elif x==9:
    return ('patrimonio_liquido')
  elif x==10:
    return ('provisoes_total')
  elif x==11:
    return ('despesas_emp')
  elif x==12:
    return ('beneficiarios')
#função para verificar a variável que o usuário escolheu e definir o nome que irá aparecer nos eixos dos gráficos
def eixo(a):
  #estrutura condicional para verificar o valor inserido pelo usuário
  if a==1:
    return ('Ativo total')
  elif a==2:
    return ('Ativo circulante')
  elif a==3:
    return ('Ativo não circulante')
  elif a==4:
    return ('Passivo total')
  elif a==5:
    return ('Passivo circulante')
  elif a==6:
    return ('Passivo não circulante')
  elif a==7:
    return ('Receita total')
  elif a==8:
    return ('Despesa total')
  elif a==9:
    return ('Patrimônio líquido')
  elif a==10:
    return ('Provisões técnicas')
  elif a==11:
    return ('Despesas com empregados')
  elif a == 12:
    return ('Beneficiários')
# função retorna número de beneficiarios cada modalidade de operadora
def soma(lista):
  som = 0 
  for i in range (len(lista)):
    som += lista[i]
  return som        
```

###Substituição dos valor em branco por zero 

```
#estrutura de repetição para percorrer todos as colunas do conjunto de dados e substituir os lugares que tem "NaN" pelo número 0
for i in range(len(dados1.columns)):
  a = dados1.columns[i]
  dados1.update(dados1[a].fillna(0))
#Estrutura de repetição para percorrer todas as colunas do banco de dados e salvar o nome da coluna 
for i in range(len(dados1.columns)):
  hh = dados1.columns[i]
  #Estrutura de repetição para percorrer todos os dados da coluna salva na estrutura anterior
  for j in range(len(dados1)):
    #Estrutura condicional para verificar se os valores da coluna é "inf" e se for substituir por 0
    if dados1[hh][j] == float('inf'):
      dados1[hh][j] = 0  
```
